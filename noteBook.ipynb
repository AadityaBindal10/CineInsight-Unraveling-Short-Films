```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
netflix_df = pd.read_csv('netflix_data.csv')
netflix_filter = netflix_df["type"] != "TV Show"
netflix_subset = netflix_df[netflix_filter]
netflix_movies = netflix_subset.loc[:,["title", "country", "genre", "release_year", "duration"]] # Removed the unnecessary colon
short_movies = netflix_movies[netflix_movies["duration"] < 60]
colors = []
for lab, row in netflix_movies.iterrows():
    if(row["genre"] == "Children"):
        colors.append('#00008B') # blue
    elif(row["genre"] == "Documentaries"):
        colors.append('#8B0000') # red
    elif(row["genre"] == "Stand-Up"):
        colors.append('#006400') # green 
    else:
        colors.append('#FF8C00') # orange
fig = plt.figure(figsize=(12, 8)) # Changed the argument to figsize
plt.scatter(y=netflix_movies["duration"], x=netflix_movies["release_year"], c=colors) # Added x and y arguments
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()
```
